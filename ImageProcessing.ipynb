```python
capture = cv2.VideoCapture('video_1.mp4')

# 0 - default camera of laptop/system
# capture = cv2.VideoCapture(0)
i = 0
while True:
    # reads one frame at a time
    flag, frame = capture.read()
    if flag:
        i += 1
        frame = cv2.resize(frame,None,fx=0.5,fy=0.5)
        # show frame on output window
        cv2.imshow("result",frame)
        cv2.imwrite("images/img_{}.png".format(i), frame)
        # wait infinitely unitl we press ESCAPE Key
        #27 - ASCII of escape
        if cv2.waitKey(1) == 27:
            break
    else:
        print("No camera detected...")
        
# close all the windows
cv2.destroyAllWindows()
# release the camera object
capture.release()
```
